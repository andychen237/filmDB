```python
import pandas as pd
```

```python
data = ['movie','directedby','starring','keywords','ranking','genre','shootinglocation']
ids = ['tt0034583','tt0035446','tt0039689','tt0042593','tt0050825','tt0053291','tt0053604','tt0062136','tt0064116','tt0074958','tt0078788','tt0087884','tt0112913','tt0166924','tt0244316']
for i in data:
    infile = i + '.txt'
    outfile = i + '-small.txt'
    source = pd.read_csv(infile, sep='\t')
    test = pd.DataFrame(columns=source.columns)
    for j in range(len(ids)):
        test = test.append(source[source.movieID == ids[j]])
    if i == 'movie':
        test['year'] = test['year'].astype(int)
        test['runtime'] = test['runtime'].astype(int)
    if i == 'shootinglocation':
        test = test.dropna()
    test.to_csv(outfile, sep='\t', index=False)
    display(test)
```

```python
starring = pd.read_csv('starring-small.txt', sep='\t')
person = pd.read_csv('person.txt', sep='\t')
person_small = pd.DataFrame(columns=person.columns)
for i in starring['personID'].values:
    person_small = person_small.append(person[person.personID == i])
person_small = person_small.drop_duplicates()
person_small = person_small.dropna()
person_small['birthYear'] = person_small['birthYear'].astype(int)
person_small.to_csv('person-small.txt', sep='\t', index=False)
display(person_small)
```
